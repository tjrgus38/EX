# EX2-11 손글씨

```python
# 필요한 모듈 import하기 

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 손글씨 데이터 준비
digits = load_digits()
digits_data = digits.data
digits_label = digits.target
```

```python
digits.keys()

# digits 안의 종류 확인
```

```python
print(digits_data.shape)

# 몇 행, 몇 열의 데이터인지 확인
```

```python
print(digits_label.shape)
digits_label
```

```python
digits.target_names
```

```python
digits.feature_names

# digits 안에 pixel이라는 파일들이 담겨져 있다.
```

```python
print(digits.DESCR)

# 8x8의 정수 형태의 이미지가 담겨져 있고, E. Alpaydin 이라는 사람이 만들었다.
# 날짜는 July 1998년등의 데이터에대한 설명이 적혀져 있다.
```

```python
import pandas as pd
print(pd.__version__)

# pandas를 불러오고 버전 확인
```

```python
digits_df = pd.DataFrame(data = digits_data, columns = digits.feature_names)
digits_df

# 확인이 쉬운 표로 정리
```

```python
digits_df[""] = digits.target
digits_df

# 마지막 열에 target을 추가 -> 65개 열로 
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# 학습, 테스트할 데이터들을 분리(순차대로 분리를 하는 것보다 섞어주는 것이 좋음)
# 테스트할 데이터의 숫자를 전체 데이터의 20%를 사용
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
# 분리한 데이터를 fit을 통해 학습을 시키고 정답률을 decision_tree로 예측
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
# 의사결정트리의 정확도는 0.855
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
# RandomForest로 예측
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
# RandomForest는 0.9638의 정확도
```

# svm

```python
from sklearn import svm
svm_model = svm.SVC()

print(svm_model._estimator_type)
```

```python
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
# svm_model의 정확도는 0.988
```

# SGD Classfier

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)
```

```python
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
# SGD classifier의 정확도는 0.939
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)
```

```python
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
# Logistic Regression의 정확도는 0.957
```

### test_size = 0.2 에서의 
#### Decision Tree의 정확도 : 0.855
#### Random Forest의 정확도 : 0.964
#### SVM의 정확도 : 0.988
#### SGD Classfier의 정확도 : 0.939
#### Logistic Regression의 정확도 : 0.957

*************
## **같은 조건에서 svm의 정확도가 가장 높았다**

```python
# 필요한 모듈 import하기
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 손글씨 데이터 준비
wine = load_wine()
wine_data = wine.data
wine_label = wine.target
```

```python
wine.keys()

# wine 안의 종류 확인
```

# EX2-13 유방암 여부

```python
# 필요한 모듈 import하기
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 유방암 데이터 준비
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target
```

```python
breast_cancer.keys()

# breast_cancer 안의 종류 확인
```

```python
print(breast_cancer_data.shape)

# 몇 행, 몇 열의 데이터인지 확인
```

```python
print(breast_cancer_label.shape)
breast_cancer_label
```

```python
breast_cancer.target_names
```

```python
breast_cancer.feature_names

# breast_cancer 안에 다양한 정보가 담겨져 있다.
```

```python
print(breast_cancer.DESCR)

# 30가지의 수치로, 악성과 비악성의 Malignant, Benign으로 분류.
# November, 1995년 Dr. William H. Wolberg, W. Nick Street, Olvi L. Mangasarian가 만들었다. 
```

```python
import pandas as pd
breast_cancer_df = pd.DataFrame(data = breast_cancer_data, columns = breast_cancer.feature_names)
breast_cancer_df

# 확인이 쉬운 표로 정리
```

```python
breast_cancer_df[""] = breast_cancer.target
breast_cancer_df

# 마지막 열에 target을 추가 -> 31개 열로 
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# 학습, 테스트할 데이터들을 분리
# 테스트할 데이터의 숫자를 전체 데이터의 20%를 사용
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
# 분리한 데이터를 fit을 통해 학습을 시키고 정답률을 decision_tree로 예측

accuracy = accuracy_score(y_test, y_pred)
accuracy
# 의사결정트리의 정확도는 0.912
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
# RandomForest로 예측
accuracy = accuracy_score(y_test, y_pred)
accuracy
# RandomForest는 1.0의 정확도
```

# svm

```python
from sklearn import svm
svm_model = svm.SVC()

print(svm_model._estimator_type)

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
# svm_model로 예측

print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
accuracy
# svm_model의 정확도는 0.903
```

# SGD Classfier

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
# SGD classifier로 예측
accuracy = accuracy_score(y_test, y_pred)
accuracy
# SGD classifier의 정확도는 0.746
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
# Logistic Regression로 예측
accuracy = accuracy_score(y_test, y_pred)
accuracy
# Logistic Regression의 정확도는 0.939
```

### test_size = 0.2 에서의

#### Random Forest의 정확도 : 1.0
